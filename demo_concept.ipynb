#Bridge Pattern

```
class Data:
    """
    Data Store Class
    list_flow_id = {"Flow_1": ["Activity_1", "Gateway_1"],
                "Flow_2": ["Gateway_1", "Activity_2"],
                "Flow_3": ["Gateway_1", "Activity_3"]}
    """

    def __init__(self):
        self.flow_sequence = 0
        self.list_flow_id = {'flow': []}

    def __get__(self, obj, klas):
        self.list_flow_id["flow"].append({
            "Flow_1": ["Activity_1", "Gateway_1"],
            "Flow_2": ["Gateway_1", "Activity_2"],
            "Flow_3": ["Gateway_1", "Activity_3"]
        })
        return {'flow': self.list_flow_id}

class BusinessLogic:
    """ Business logic holding data store instances """

    data = Data()

    def flow_list(self):
        return self.data['flow'].keys()

    def flow_list_information(self, flow_id):
        return self.data['flow'].get(flow_id, None)

class Ui:
    """ UI interaction class """

    def __init__(self):
        self.business_logic = BusinessLogic()

    def get_flow_information(self, flow):
        flow_info = self.business_logic.flow_list_information(flow)
        if flow_info:
            print('PRODUCT INFORMATION:', flow_info)
        else:
            print('That flow does not exist in the records'.format(flow))

def main():
    ui = Ui()
    ui.get_flow_information('flow')

if __name__ == '__main__':
    main()
```

```
import spacy
import uuid
import json
import xml.etree.ElementTree as gfg
from json import JSONEncoder, loads, dumps
from xml.etree import ElementTree

# ==================================================================================
#                                  CREATE JSON
# ==================================================================================
try:
    from types import SimpleNamespace as Namespace
except ImportError:
    from argparse import Namespace

class Object:
    """
    method : toJSON
    :return: dumps JSON
    """
    def toJSON(self):
        return dumps(self, default=lambda o: o.__dict__, sort_keys=True, indent=4, cls=BPMNEncoder)

class DumpDict(dict):
    """
    :return: json.dumps(self)
    """
    def __str__(self):
        return json.dumps(self)

class BPMNEncoder(JSONEncoder):
    """
    method : default
    """
    def default(self, o): return o.__dict__

class json_dictionary(dict):
    # __init__ function
    def __init__(self):
        self = dict()

    # Function to add key:value
    def add(self, key, value):
        self[key] = value

# ==================================================================================
#                                      NLP
# ==================================================================================
class NLP():
    def __init__(self, text):
        self.word_active = []
        self.text = text
        self.nlp = spacy.load("en_core_web_sm")
        self.doc = self.nlp(self.text)
        self.subjects = []
        self.verbs = []
        self.objects = []
        self.sentence_order = 0
        self.verb_counter = 0
        self.verb_order = 0
        self.prepare_flow_list = {'flow_id': []}

    def merge_phrases(self):
        """
        Merge Phrase
        :return: self.doc
        """
        with self.doc.retokenize() as retokenizer:
            for np in list(self.doc.noun_chunks):
                attrs = {
                    "tag": np.root.tag_,
                    "lemma": np.root.lemma_,
                    "ent_type": np.root.ent_type_,
                }
                retokenizer.merge(np, attrs=attrs)
        return self.doc

    def merge_punct(self):
        """
        merge_punct
        :return: self.doc
        """
        spans = []
        for word in self.doc[:-1]:
            if word.is_punct or not word.nbor(1).is_punct:
                continue
            start = word.i
            end = word.i + 1
            while end < len(self.doc) and self.doc[end].is_punct:
                end += 1
            span = self.doc[start:end]
            spans.append((span, word.tag_, word.lemma_, word.ent_type_))
        with self.doc.retokenize() as retokenizer:
            for span, tag, lemma, ent_type in spans:
                attrs = {"tag": tag, "lemma": lemma, "ent_type": ent_type}
                retokenizer.merge(span, attrs=attrs)
        return self.doc

    def find_dobj(self, token):
        """
        find_dobj
        :param token:
        :return: dobj
        """
        dobj = [tok for tok in token.rights if tok.dep_ == 'dobj']
        return dobj

    def find_role_prep_pobj(self, token):
        """
        # ROLE # dobj with prep to pobj
        :param token:
        :return: p_obj, agent
        """
        try:
            j = token.i + 1
            if self.doc[j + 1].dep_ == "prep" or self.doc[j + 1].text in ["to"]:
                p_obj = self.doc[j].text
                return p_obj
            elif self.doc[j].dep_ == "agent":
                agent = self.doc[j + 1].text
                return agent
        except IndexError:
            return None

    def find_role_subject_prep_pobj(self, token):
        """
        find_role_subject_prep_pobj
        :param token:
        :return: subj + " " + prep + " " + str(pobj)
        """
        subj = token.text
        i = token.i + 1
        prep = self.doc[i].text
        if self.doc[i].dep_ == "prep":
            pobj = self.doc[i + 1].text
            return subj + " " + prep + " " + str(pobj)

    def find_role_dobj_prep_pobj(self, token):
        """
        find_role_dobj_prep_pobj
        :param token:
        :return: prep + " " + str(pobj)
        """
        try:
            i = token.i + 1
            prep = self.doc[i].text
            if self.doc[i].dep_ == "prep":
                pobj = self.doc[i + 1].text
                return prep + " " + str(pobj)
        except IndexError:
            return None

    def find_mark_subjectpass(self, token):
        """
        find_mark_subjectpass
        :param token:
        :return: nsubjpass
        """
        # Mark
        if token.dep_ == 'mark':
            mark = token.text
            print("mark", mark)
            i = token.i + 1
            dep = self.doc[i].dep_
            if dep in ["nsubjpass"]:
                nsubjpass = self.doc[i].text
                return nsubjpass

    def find_role_subject(self, token):
        """
        find_role_subject
        :param token:
        :return: nsubj + " " + prep + " " + str(p_obj)
        """
        nsubj = token.text
        i = token.i + 1
        prep = self.doc[i].text
        if self.doc[i].dep_ == "prep":
            p_obj = self.doc[i + 1].text
            return nsubj + " " + prep + " " + str(p_obj)

    def find_role_subjectpass(self, token):
        """
        find_role_subjectpass
        :param token:
        :return: nsubjpass + " " + prep + " " + str(p_obj)
        """
        if token.dep_ == 'nsubjpass':
            nsubjpass = token.text
            i = token.i + 1
            prep = self.doc[i].text
            if self.doc[i].dep_ == "prep":
                p_obj = self.doc[i + 1].text
                return nsubjpass + " " + prep + " " + str(p_obj)

    def find_subjpass(self, token, doc):
        """
        find_subjpass
        :param token:
        :param doc:
        :return: str(doc[i].text)
        """
        i = token.i - 1
        if doc[i].pos_ in ["NOUN", "PROPN"]:
            return str(doc[i].text)

    def find_activity(self, verb, dobj):
        """
        find_activity
        :param verb:
        :param dobj:
        :return: Activity_1_xxx
        """
        return "Activity_" + str(self.verb_order) + "_" + verb + "_" + str(dobj).replace('[', '').replace(']', '')

    def find_gateway(self, token, doc):
        """
        find_gateway
        :param token:
        :param doc:
        :return: Gateway_1_xxx
        """
        try:
            i = token.i + 1
            if doc[i].dep_ in ["cc"] and doc[i + 1].pos_ in ["VERB"]:
                return "Gateway_"+str(self.verb_order)+"_"+doc[i].text
            elif doc[i].dep_ in ["cc"]:
                return "Gateway_"+str(self.verb_order)+"_"+doc[i].text
        except IndexError:
            return None

    def find_relation_of_verb_active(self):
        """
        Relation of VERB
        :return: self.word_active
        """
        doc = self.merge_phrases()
        doc = self.merge_punct()

        for sent in enumerate(doc.sents):
            print([sent[i] for i in range(len(sent))])
            self.sentence_order += 1

            for token in doc:
                if token.pos_ == 'VERB':
                    subj = self.find_subjpass(token, doc)
                    subjpass = self.find_mark_subjectpass(token)
                    dobj = self.find_dobj(token)
                    role = self.find_role_prep_pobj(token)

                    # Activity_1_xxx", "Gateway_1_yyy", "Activity_2_zzz", "Activity_3_aaa
                    verb = token.text
                    if verb not in ["can"]:
                        self.verb_order += 1

                        activity = self.find_activity(verb, dobj)
                        if activity is not None:
                            self.word_active.append(str(activity)+"|")

                        gateway = self.find_gateway(token, doc)
                        if gateway is not None:
                            self.word_active.append(str(gateway)+"|")
                else:
                    if token.dep_ == 'nsubj':
                        role = self.find_role_subject_prep_pobj(token)
                    elif token.dep_ == 'dobj':
                        role = self.find_role_dobj_prep_pobj(token)
        print(self.word_active)
        print("============END NLP=================")

    def prepare_activity(self):
        """
        # step 1
        seq_1 = ["Activity_1", "Gxs"]
        seq_2 = ["Gxs", "Activity_2"]
        seq_3 = ["Gxs", "Activity_3"]
        :return: json_dict
        """
        json_dict = json_dictionary()
        #data = ["Register Leave Request|", "Gxs.|", "Verify available vacation days|", "Approve leave request|"]  # task ให้ split string แล้วเช็ค Activity
        data = self.word_active
        result = ([x + y for x, y in zip(data, data[1:] + data[:0])])
        # print(result[0][:-1])

        re = []
        for ls in result:
            re.append(ls[:-1])

        #composed data
        x = []
        for (i, item) in enumerate(re):
            a = (i, str(item).split('|'))
            x.append(a)

        #generate flow id
        for index, value in x:
            print(index, value)
            json_dict.add("Flow_"+str(uuid.uuid4()), value)
        return json_dict

    #BPMN
    def bpmn_process(self, event_id):
        """
        <bpmn:process id="Process_0eui6ju" isExecutable="true">
            <bpmn:task id="Activity_1e1d04r">
                <bpmn:outgoing>Flow_1povr4d</bpmn:outgoing>
            </bpmn:task>
            <bpmn:task id="Activity_1idii86">
                <bpmn:incoming>Flow_1povr4d</bpmn:incoming>
            </bpmn:task>
            <bpmn:sequenceFlow id="Flow_1povr4d" sourceRef="Activity_1e1d04r" targetRef="Activity_1idii86" />
        </bpmn:process>

        :param task_name:
        :param event_id:
        :return: element
        """
        # start process
        print("================START PROCESS BPMN====================")
        x = []
        element = gfg.Element("bpmn:process", id=event_id, isExecutable="true")
        list_flow_id = self.prepare_activity()
        for key, value in list_flow_id.items():
            #incoming = 0, outgoing = 1
            print("Process key->", key, value[0], value[1])
            in_coming = str(value[0])
            out_going = str(value[1])
            flow_bpmn_element = str(key)

            in_coming_uuid = ""
            out_going_uuid = ""

            if str(in_coming).find("Start") == 0:
                in_coming_uuid = "StartEvent_1"
            elif str(in_coming).find("Activity") == 0:
                in_coming_uuid = "Activity_"+str(uuid.uuid4())
            elif str(in_coming).find("Gateway") == 0:
                in_coming_uuid = "Gateway_"+str(uuid.uuid4())
            elif str(out_going).find("Activity") == 0:
                out_going_uuid = "Activity_" + str(uuid.uuid4())
            elif str(out_going).find("Gateway") == 0:
                out_going_uuid = "Gateway_"+str(uuid.uuid4())

            if len(in_coming) > 0 or len(out_going) > 0:
                # task
                task = self.bpmn_task(in_coming=in_coming, out_going=out_going, flow_id=key,
                                      in_coming_uuid=in_coming_uuid, out_going_uuid=out_going_uuid, task_name="test")
                element.insert(1, task)

            # sequence
            sequence = self.bpmn_sequence_flow(flow_bpmn_element, in_coming, out_going)
            element.insert(1, sequence)

        print("Result Process =>", ElementTree.tostring(element).decode("utf-8"))
        return element

    def bpmn_task(self, in_coming, out_going, flow_id, in_coming_uuid, out_going_uuid, task_name):
        """
        <bpmn:task id="Activity_1smlx8n" name="B">
          <bpmn:outgoing>Flow_000nznv</bpmn:outgoing>
        </bpmn:task>
        :param task_id:
        :param flow_id:
        :param task_name:
        :return: element
        """
        if str(in_coming).find("Activity") == 0:
            element = gfg.Element("bpmn:task", id=in_coming_uuid, name=task_name)
            activity, gateway = self.prepare_source_in_coming(in_coming, flow_id)
            gfg.SubElement(element, "bpmn:incoming").text = activity
            return element
        elif str(in_coming).find("Activity") == 0:
            element = gfg.Element("bpmn:task", id=in_coming_uuid, name=task_name)
            activity, gateway = self.prepare_source_in_coming(in_coming, flow_id)
            gfg.SubElement(element, "bpmn:incoming").text = activity
            return element
        elif str(in_coming).find("Gateway") == 0:
            element = gfg.Element("bpmn:task", id=in_coming_uuid, name=task_name)
            activity, gateway = self.prepare_source_in_coming(in_coming, flow_id)
            gfg.SubElement(element, "bpmn:incoming").text = gateway
            return element
        elif str(out_going).find("Activity") == 0:
            element = gfg.Element("bpmn:task", id=out_going_uuid, name=task_name)
            activity, gateway = self.prepare_source_out_going(in_coming, flow_id)
            gfg.SubElement(element, "bpmn:outgoing").text = activity
            return element
        elif str(out_going).find("Gateway") == 0:
            element = gfg.Element("bpmn:task", id=out_going_uuid, name=task_name)
            activity, gateway = self.prepare_source_out_going(in_coming, flow_id)
            gfg.SubElement(element, "bpmn:outgoing").text = gateway
            return element

    def bpmn_gateway(self, task_id, flow_id, task_name):
        """
        <bpmn:exclusiveGateway id="Gateway_036n3qz">
          <bpmn:incoming>Flow_000nznv</bpmn:incoming>
          <bpmn:outgoing>Flow_1gmw0cn</bpmn:outgoing>
          <bpmn:outgoing>Flow_17knxn0</bpmn:outgoing>
        </bpmn:exclusiveGateway>
        :param task_id:
        :param flow_id:
        :param task_name:
        :return: element
        """
        element = gfg.Element("bpmn:exclusiveGateway", id=task_id, name=task_name)
        gfg.SubElement(element, "bpmn:incoming").text = flow_id
        return element

    def bpmn_sequence_flow(self, event_id, source_ref, target_ref):
        """
        <bpmn:sequenceFlow id="Flow_1povr4d" sourceRef="Activity_1e1d04r" targetRef="Activity_1idii86" />
        :param event_id:
        :param source_ref:
        :param target_ref:
        :return: element
        """
        element = gfg.Element("bpmn:sequenceFlow", id=event_id, sourceRef=source_ref, targetRef=target_ref)
        return element

    def prepare_source_in_coming(self, source, flow_id):
        """
        prepare_source_in_coming
        :param source:
        :param flow_id:
        :return: in_coming_activity, in_coming_gateway
        """
        in_coming_activity = None
        in_coming_gateway = None
        if str(source).find("Activity") == 0:
            in_coming_activity = flow_id
        elif str(source).find("Gateway") == 0:
            in_coming_gateway = flow_id
        return in_coming_activity, in_coming_gateway

    def prepare_source_out_going(self, target, flow_id):
        """
        prepare_source_out_going
        :param target:
        :param flow_id:
        :return: out_going_activity, out_going_gateway
        """
        out_going_activity = None
        out_going_gateway = None

        if str(target).find("Activity") == 0:
            out_going_activity = flow_id
        elif str(target).find("Gateway") == 0:
            out_going_gateway = flow_id
        return out_going_activity, out_going_gateway

    #Diagram
    def bpmn_diagram_plan_start(self, event_id):
        """
        <bpmndi:BPMNDiagram id="BPMNDiagram_1">
        :param event_id:
        :return: element
        """
        element = gfg.Element("bpmndi:BPMNDiagram", id=event_id)
        return element

    def bpmn_diagram_plan(self, event_id, uuid):
        """
        <bpmndi:BPMNPlane id="BPMNPlane_1" bpmnElement="Process_1lica4e">
        :param event_id:
        :return: element
        """
        element = gfg.Element("bpmndi:BPMNPlane", id=event_id, bpmnElement=uuid)
        return element

    def bpmn_diagram(self, diagram_id, plan_id, uuid):
        """
        # Start => VP  | VP => Gateway  | VP => End

        list_flow_id = {"Flow_1": ["Activity_1", "Gx"],
                        "Flow_2": ["Gx", "Activity_2"],
                        "Flow_3": ["Gx", "Activity_3"]}

        <dc:Bounds x="160" y="77" width="100" height="80" />
        +100 = 260, +30 = 117
        <di:waypoint x="260" y="117" />
        +50 = 310, + 30 = 117
        <di:waypoint x="310" y="117" />

        :param event_id:
        :param task:
        :return: element

        <bpmndi:BPMNDiagram id="BPMNDiagram_1">
            <bpmndi:BPMNPlane id="BPMNPlane_1" bpmnElement="Process_1ogzzd6">
              <!-- START -->
              <bpmndi:BPMNShape id="Activity_1njqlj1_di" bpmnElement="Activity_1njqlj1">
                <dc:Bounds x="160" y="80" width="100" height="80" />
              </bpmndi:BPMNShape>
              <bpmndi:BPMNShape id="Gateway_1lh8ot4_di" bpmnElement="Gateway_1lh8ot4" isMarkerVisible="true">
                <dc:Bounds x="315" y="95" width="50" height="50" />
              </bpmndi:BPMNShape>
              <bpmndi:BPMNShape id="Activity_1hligz9_di" bpmnElement="Activity_1hligz9">
                <dc:Bounds x="420" y="80" width="100" height="80" />
              </bpmndi:BPMNShape>

              <!-- START -->
              <bpmndi:BPMNEdge id="Flow_0u8am50_di" bpmnElement="Flow_0u8am50">
                <di:waypoint x="260" y="120" />
                <di:waypoint x="315" y="120" />
              </bpmndi:BPMNEdge>
              <bpmndi:BPMNEdge id="Flow_0570dg1_di" bpmnElement="Flow_0570dg1">
                <di:waypoint x="365" y="120" />
                <di:waypoint x="420" y="120" />
              </bpmndi:BPMNEdge>

            </bpmndi:BPMNPlane>
        </bpmndi:BPMNDiagram>

        -diagram
            -plan
                -edge
                -shape
            -plan
        -diagram
        """

        print("---------------START DIAGRAM-------------------")
        diagram = self.bpmn_diagram_plan_start(diagram_id)
        plan = self.bpmn_diagram_plan(plan_id, uuid=uuid)
        diagram.append(plan)

        # XML Waypoint Start
        start_x = 160
        start_y = 80

        list_flow_id = self.prepare_activity()
        for key, value in list_flow_id.items():
            print("Diagram key->", key, value[0], value[1])

            in_coming = str(value[0])
            event_id_incoming = str(value[0])
            event_id_di_incoming = str(value[0])+"_di"

            out_going = str(value[1])
            event_id_outgoing = str(value[1])
            event_id_di_outgoing = str(value[1])+"_di"
            print("out=>", event_id_di_outgoing)

            flow_bpmn_element = str(key)
            flow_id_element = str(key)+"_di"
            print("Flow=>", flow_id_element)
            print("=========================================")

            # XML Edge Start
            edge = gfg.Element("bpmndi:BPMNEdge", id=flow_id_element, bpmnElement=flow_bpmn_element)

            # XML Shape
            if str(in_coming).find("Start") == 0:
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_incoming, bpmnElement=event_id_di_incoming)
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x), y=str(start_y), width="36", height="36")
                plan.append(shape)

                # XML Waypoint
                print("Start in_coming calculation X1, Y1")
                waypoint = self.prepare_calculate_start_event_waypoint_x1(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            if str(in_coming).find("Activity") == 0:
                # XML Shape
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_di_incoming, bpmnElement=event_id_incoming)
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x), y=str(start_y), width="100", height="80")
                plan.append(shape)

                # XML Waypoint
                print("Activity in_coming calculation X1, Y1")
                waypoint = self.prepare_calculate_waypoint_x1(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            if str(out_going).find("Activity") == 0:
                # XML Shape
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_di_outgoing, bpmnElement=event_id_outgoing)
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x), y=str(start_y), width="100", height="80")
                plan.append(shape)

                # XML Waypoint
                print("Activity out_going calculation 2X, 2Y")
                waypoint = self.prepare_calculate_waypoint_x2(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            if str(in_coming).find("Gateway") == 0:
                # XML Shape
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_di_incoming, bpmnElement=event_id_incoming, isMarkerVisible="true")
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x + 55), y=str(start_y + 15), width="50", height="50")
                plan.append(shape)

                # XML Waypoint
                print("Gateway in_coming calculation X1, Y1")
                waypoint = self.prepare_calculate_waypoint_x1(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            if str(out_going).find("Gateway") == 0:
                # XML Shape
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_di_outgoing, bpmnElement=event_id_outgoing, isMarkerVisible="true")
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x + 55), y=str(start_y + 15), width="50", height="50")
                plan.append(shape)

                # XML Waypoint
                print("Gateway out_going calculation 2X, 2Y")
                waypoint = self.prepare_calculate_waypoint_x2(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            if str(out_going).find("End") == 0:
                # end
                # XML Shape
                shape = gfg.Element("bpmndi:BPMNShape", id=event_id_di_outgoing, bpmnElement=event_id_outgoing)
                gfg.SubElement(shape, "dc:Bounds", x=str(start_x + 52), y=str(start_y + 42), width="36", height="36")
                plan.append(shape)

                # XML Waypoint
                print("End out_going calculation 2X, 2Y")
                waypoint = self.prepare_calculate_waypoint_x2(start_x=start_x, start_y=start_y)
                edge.insert(1, waypoint)

            #plan append all edge
            plan.append(edge)

        # End
        print("Result Diagram =>", ElementTree.tostring(diagram).decode("utf-8"))

    def prepare_calculate_start(self, start_x, start_y):
        """
        # x = 112 + (36) = 148
        # y = 102 + (18) = 120
        :param start_x:
        :param start_y:
        :return: waypoint
        """
        start_x += 36
        start_y += 18
        waypoint = self.bpmn_diagram_waypoint(start_x, start_y)
        return waypoint

    def prepare_calculate_start_event_waypoint_x1(self, start_x, start_y):
        """
        # x = 160 + (36) = 196
        # y = 80 + (18) = 98
        :param start_x:
        :param start_y:
        :return: waypoint
        """
        start_x += 36
        start_y += 18
        waypoint = self.bpmn_diagram_waypoint(start_x, start_y)
        return waypoint

    def prepare_calculate_start_event_waypoint_x2(self, start_x, start_y):
        """
        # x = 196 + (55) = 261
        # y = 80 + (18) = 98
        :param start_x:
        :param start_y:
        :return: waypoint
        """
        start_x += 55
        start_y += 18
        waypoint = self.bpmn_diagram_waypoint(start_x, start_y)
        return waypoint

    def prepare_calculate_waypoint_x1(self, start_x, start_y):
        """
        # x = 160 + (100) = 260
        # y = 80 + (40) = 120
        :param start_x:
        :param start_y:
        :return: waypoint
        """
        start_x += 100
        start_y += 40
        waypoint = self.bpmn_diagram_waypoint(str(start_x), str(start_y))
        return waypoint

    def prepare_calculate_waypoint_x2(self, start_x, start_y):
        """
        # x = 160 + (100) = 260
        # y = 80 + (40) = 120
        :param start_x:
        :param start_y:
        :return: waypoint
        """
        start_x += 50
        start_y += 40
        waypoint = self.bpmn_diagram_waypoint(start_x, start_y)
        return waypoint

    def bpmn_diagram_waypoint(self, x, y):
        """
        bpmn_diagram_waypoint
        :param x:
        :param y:
        :return: waypoint
        """
        waypoint = gfg.Element("di:waypoint", x=str(x), y=str(y))
        return waypoint

if __name__ == "__main__":
    nlp = spacy.load("en_core_web_sm")
    #text = "After the payment is confirmed, the process ends."
    text = "A member of the sales department can then reject or accept the order for a customized bike."
    uuid_process = "Process_"+str(uuid.uuid4())
    custom = NLP(text)
    custom.find_relation_of_verb_active()
    custom.prepare_activity()
    custom.bpmn_process(event_id=uuid_process)
    custom.bpmn_diagram(diagram_id="BPMNDiagram_1", plan_id="BPMNPlane_1", uuid=uuid_process)
```
