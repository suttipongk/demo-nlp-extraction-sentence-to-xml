```
import spacy.utils.nltk
from spacy.utils.nltk.corpus import wordnet
from spacy.utils.nltk.stem import WordNetLemmatizer

def get_wordnet_pos(word):
    tag = spacy.utils.nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

lemmatizer = WordNetLemmatizer()
word = 'feet'
print(lemmatizer.lemmatize(word, get_wordnet_pos(word)))

sentence = "The striped bats are hanging on their feet for best"
print([lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in spacy.utils.nltk.word_tokenize(sentence)])
```
