```
from nltk_test.corpus import stopwords
print(stopwords.words('english'))

def content_fraction(text):
    stopwords = nltk_test.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

print(content_fraction(nltk_test.corpus.reuters.words()))

puzzle_letters = nltk_test.FreqDist('egivrvonl')
obligatory = 'r'
wordlist = nltk_test.corpus.words.words()
print([w for w in wordlist if len(w) >= 6 and obligatory in w and nltk_test.FreqDist(w) <= puzzle_letters])

from nltk_test.corpus import wordnet as wn
print(wn.synsets('motorcar')[0])
print(wn.synsets('motorcar')[0].definition())
for synset in wn.synsets('car'):
    print (synset.lemma_names)
```
